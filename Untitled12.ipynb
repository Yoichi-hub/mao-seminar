```python
import pandas as pd
import io
import requests
import numpy as np
import time
import datetime
import math
from datetime import timedelta
import matplotlib as mpl
import matplotlib.pyplot as plt


url="https://www3.nhk.or.jp/n-data/opendata/coronavirus/nhk_news_covid19_prefectures_daily_data.csv"
s=requests.get(url).content
df=pd.read_csv(io.StringIO(s.decode('utf-8')))
df["date"]=pd.to_datetime(df["日付"] )
df.set_index('date')
df=df[df.columns[[8,2,3,5]]]
df=df.rename(columns={'都道府県名':'kenName','各地の感染者数_1日ごとの発表数':'dayNumber','各地の死者数_1日ごとの発表数':'dayDeaths'})
df
```

```python
def func(todouhuken):
    tokyo=df[df['kenName'].str.contains(todouhuken)]
    tokyo1=tokyo.shift(5)
    tokyo1=tokyo1.rename(columns={'dayNumber':'dayDelayNumber','dayDeaths':'dayDelayDeaths'})
    tokyo1['date']=tokyo1['date']+datetime.timedelta(days=5)
    tokyo2=pd.merge(tokyo,tokyo1,on="date")
    tokyo2["differ"]=tokyo2["dayNumber"]-tokyo2["dayDelayNumber"]
    tokyo2["differ_ratio"]=(tokyo2["dayNumber"]-tokyo2["dayDelayNumber"])/tokyo2["dayDelayNumber"]
    
    x=tokyo2['date']
    y=tokyo2['dayNumber']
    z=tokyo2['differ_ratio']

    l1="dayNumbers"
    l2="different ratio/5days"

    fig=plt.figure()
    ax1=fig.add_subplot(2,1,1)
    ax2=fig.add_subplot(2,1,2)

    ax1.set_yscale("log")

    ax1.plot(x,y,color="red",label=l1)
    ax2.plot(x,z,color="green", label=l2)


func('東京')
```
